# Graph Clustering to Detect Anomalies

```python
edgelist = df[["callgno","calldno"]]
edgelist = edgelist.astype(str)
edgelist.to_csv("edgelist.csv",index=False)
G = nx.read_edgelist("edgelist.csv",delimiter=',')
```

```python
#print df[df['calldno']==Anomalies[0]]
#df.calldno=="8291819583"
tmp = nx.Graph()
for i in range(0,len(Anomalies)):
    tmp = nx.compose(tmp,nx.ego_graph(G,Anomalies[i],radius=3))

len(tmp.nodes())
```

```python
pagerank_matrix = nx.pagerank(tmp,max_iter=200)
pagerank_matrix_df = pd.DataFrame.from_dict(pagerank_matrix,orient='index')
pagerank_matrix_df.column=["Page Rank"]
pagerank_matrix_df.describe()
```

```python
trianglecount_matrix = nx.triangles(tmp)
trianglecount_matrix_df = pd.DataFrame.from_dict(trianglecount_matrix,orient='index')
trianglecount_matrix_df.column=["Triangle Count"]
trianglecount_matrix_df.describe()
```

```python
# betweenness_centrality_matrix = nx.betweenness_centrality(tmp)
# betweenness_centrality_matrix_df = pd.DataFrame.from_dict(betweenness_centrality_matrix,orient='index')
```

```python
body = pd.merge(pagerank_matrix_df,trianglecount_matrix_df,right_index=True,left_index=True)
body.head()
body = body.reset_index()
body.columns = ["Phone Number", "Page Rank", "Triangle Count"]
```

```python
body.values
```

```python
from sklearn.decomposition import PCA, KernelPCA
import matplotlib.pyplot as plt
X=body.values[:,[1,2]]
pca = PCA(n_components=2)
pca.fit(X)
X_lowed = pca.transform(np.real(X))
plt.plot(pca.explained_variance_)
plt.plot(X_lowed[:,0],'o')
NumOfFrauds = len(X_lowed[X_lowed[:,0] > (3*np.std(X_lowed[:,0]))])
indices = np.argsort(X_lowed[:,0])[:len(X_lowed)-(NumOfFrauds+1):-1]
indices
SuspectFraudPhoneNumbers = body.iloc[indices]["Phone Number"].values
```

```python
X_lowed[X_lowed[:,0] > (2*np.std(X_lowed[:,0]))]
```

```python
pr_max= body["Page Rank"].max()
pr_min= body["Page Rank"].min()
tc_max= body["Triangle Count"].max()
tc_min= body["Triangle Count"].min()
print pr_max, pr_min, tc_max, tc_min
```

```python
Normal_indices = np.argsort(X_lowed[:,0])[0:len(X_lowed)-(NumOfFrauds+1)]
X_normal = body.iloc[Normal_indices][["Page Rank","Triangle Count"]].values
X_abnormal = body.iloc[indices][["Page Rank","Triangle Count"]].values
from sklearn import preprocessing
scaler = preprocessing.StandardScaler().fit(X_normal)
X_normal = scaler.transform(X_normal)
X_abnormal = scaler.transform(X_abnormal)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager
from sklearn import svm

#Normal_indices = np.argsort(X_lowed[:,0])[0:len(X_lowed)-(NumOfFrauds+1)]
#X_normal = body.iloc[Normal_indices][["Page Rank","Triangle Count"]].values
#X_normal[:,0]
#X_abnormal = body.iloc[indices][["Page Rank","Triangle Count"]].values

xx, yy = np.meshgrid(np.linspace(pr_min, pr_max, 500), np.linspace(tc_min, tc_max, 500))
# Generate train data
#X = 0.3 * np.random.randn(100, 2)
#X_train = np.r_[X + 2, X - 2]
X_train = X_normal[0:100000]
X_test = X_normal[100000:]
#X=X_normal[200000:]
# Generate some regular novel observations
#X = 0.3 * np.random.randn(20, 2)
#X_test = np.r_[X + 2, X - 2]
# Generate some abnormal novel observations
X_outliers = X_abnormal
#X_outliers = np.random.uniform(low=-4, high=4, size=(20, 2))

# fit the model
clf = svm.OneClassSVM(nu=0.1, kernel="rbf", gamma=0.1)
clf.fit(X_train)
y_pred_train = clf.predict(X_train)
y_pred_test = clf.predict(X_test)
y_pred_outliers = clf.predict(X_outliers)
n_error_train = y_pred_train[y_pred_train == -1].size
n_error_test = y_pred_test[y_pred_test == -1].size
n_error_outliers = y_pred_outliers[y_pred_outliers == 1].size

# plot the line, the points, and the nearest vectors to the plane
Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.title("Novelty Detection")
plt.contourf(xx, yy, Z, levels=np.linspace(Z.min(), 0, 70), cmap=plt.cm.Blues_r)
a = plt.contour(xx, yy, Z, levels=[0], linewidths=2, colors='red')
plt.contourf(xx, yy, Z, levels=[0, Z.max()], colors='orange')

b1 = plt.scatter(X_train[:, 0], X_train[:, 1], c='white')
b2 = plt.scatter(X_test[:, 0], X_test[:, 1], c='green')
c = plt.scatter(X_outliers[:, 0], X_outliers[:, 1], c='red')
plt.axis('tight')
plt.xlim((pr_min, pr_max))
plt.ylim((tc_min, tc_max))
plt.legend([a.collections[0], b1, b2, c],
           ["learned frontier", "training observations",
            "new regular observations", "new abnormal observations"],
           loc="upper left",
           prop=matplotlib.font_manager.FontProperties(size=11))
plt.xlabel(
    "error train: %d/200 ; errors novel regular: %d/40 ; "
    "errors novel abnormal: %d/40"
    % (n_error_train, n_error_test, n_error_outliers))
plt.show()
```

# SparK can be used on Big Data Real-time Streaming detection of Anomalies to scale at Peta Bytes

```python
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import cross_validation

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print "Running Logistic Regression ..."
clf  = RandomForestClassifier(n_estimators=3,min_samples_leaf=3)

clf = clf.fit(X_train, y_train)
scores = cross_validation.cross_val_score(clf, X, y, cv=5)
print "%s -- %s" % (clf.__class__, np.mean(scores))
probas_ = clf.predict_proba(X_test)
```

```python
from sklearn.metrics import roc_curve, auc

# Compute ROC curve and area the curve
fpr, tpr, thresholds = roc_curve(y_test, probas_[:, 1])
roc_auc = auc(fpr, tpr)
print("Area under the ROC curve : %f" % roc_auc)

# Plot ROC curve
plt.clf()
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic example')
plt.legend(loc="lower right")
plt.show()
```
